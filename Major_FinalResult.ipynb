```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define evaluation metrics for each model
models = ['LSTM', 'Random Forest', 'Decision Tree']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-score']

lstm_scores = [0.92, 0.89, 0.91, 0.90]
rf_scores = [0.87, 0.85, 0.86, 0.85]
dt_scores = [0.81, 0.79, 0.80, 0.78]

# Set up the bar width and position
bar_width = 0.25
x = np.arange(len(metrics))

# Define custom colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, Orange, Green

# Plot bars for each model
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, lstm_scores, width=bar_width, label='LSTM', color=colors[0])
plt.bar(x, rf_scores, width=bar_width, label='Random Forest', color=colors[1])
plt.bar(x + bar_width, dt_scores, width=bar_width, label='Decision Tree', color=colors[2])

# Customize graph
plt.xlabel('Evaluation Metrics')
plt.ylabel('Score')
plt.title('Model Performance Comparison')
plt.xticks(ticks=x, labels=metrics)
plt.ylim(0, 1)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
